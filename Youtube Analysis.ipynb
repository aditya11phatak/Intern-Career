```python
#Step1: Set up Environment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step2:Load and Explore Data-Type

#Load Data-Set
df=pd.read_csv("Downloads/youtubers_df.csv")
df
```

```python
#Step3: Data Cleaning

# Handling missing values (if any)|
df.dropna(inplace=True)  # Drop rows with missing values, or alternatively, use df.fillna()

# Convert columns to appropriate data types if needed
# For example: convert 'Date' column to datetime if present
# df['date_column'] = pd.to_datetime(df['date_column'])

# Verify the changes
df.info()
```

```python
#head() shows the first few rows to get a feel for the data.
#info() gives us a concise summary of the DataFrame, including data types and non-null counts.
#isnull().sum() helps to check for any missing values.
#describe() provides statistical insights into numerical columns, such as mean, median, and standard deviation.
```

```python
#Step4:Manupulating Columns Data
# Example: Create a new feature 'Engagement Rate' if the dataset has 'views', 'likes', and 'comments'
df['engagement_rate'] = (df['Likes'] + df['Comments']) / df['Visits']

# Check the new column
df[['Visits', 'Likes', 'Comments', 'engagement_rate']].head()
```

```python
#Step5: Exploratory Data Analysis (EDA)

#Distribution Of Views
#HISTOGRAM For Views
plt.figure(figsize=(10,6))
sns.histplot(df['Visits'], bins=30, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Visit')
plt.ylabel('Frequency')
plt.show()
```

```python
# BOX PLOT for engagement rate
plt.figure(figsize=(10,6))
sns.boxplot(x=df['engagement_rate'])
plt.title('Distribution of Engagement Rate')
plt.show()
```

```python
# SCATTER PLOT for views and likes
plt.figure(figsize=(10,6))
sns.scatterplot(x='Visits', y='Likes', data=df)
plt.title('Visits vs Likes')
plt.xlabel('Visits')
plt.ylabel('Likes')
plt.show()
```

```python
# Box plot for engagement rate
plt.figure(figsize=(10,6))
sns.boxplot(x=df['engagement_rate'])
plt.title('Distribution of Engagement Rate')
plt.show()
```

```python
# Correlation heatmap
# Step 1: Select only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['number'])

# Step 2: Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Step 3: Plot the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pair plot
plt.figure(figsize=(10,6))
sns.pairplot(df[['Visits', 'Likes', 'Comments', 'engagement_rate']])
plt.show()
```

```python
# If there's a categorical column, like 'category'
plt.figure(figsize=(50,20))
sns.catplot(x='Categories', y='Visits', kind='bar', data=df)
plt.title('Average Views by Categories')
plt.xticks(rotation=90)
plt.show()
```
